```python
import pandas as pd
```

```python
# reading in data
df=pd.read_csv('../ZILLOW-REGIONS.csv')
```

```python
df.head()
```

```python
# dropping uneccesary column
df=df.drop('region_type',axis=1)
```

```python
df.head()
```

```python
# initializing new column
df['zip']=''
df
```

```python
# setting new column equal to the zipcode found in the 'region' column. Removing any rows that excluded a zipcode
count=0
to_drop=[]
for i in range(len(df)):
    try:
        df.iloc[i,2]=int(df.iloc[i,1].split(';')[0])
    except:
        print(f'{df.iloc[i,1].split(";")[0]} could not be parsed as int')
        count=count+1
        to_drop.append(i)
print(count)
```

```python
df=df.drop(to_drop,axis=0)
```

```python
# resetting index to the unique 'region_id' column
df=df.set_index('region_id')
df
```

```python
# writing to a new csv
df.to_csv('../zillow_regions_parsed.csv')
```
